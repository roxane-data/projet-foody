```python
# définir la fct sql_to_pandas

def sql_to_df(sql_query):

    # Use pandas to pass sql query using connection form SQLite3
    df = pd.read_sql(sql_query, conn)

    # Show the resulting DataFrame
    return df
```

Calculer, pour la commande numéro 10251, pour chaque produit acheté dans celle-ci, le

Afficher donc (dans une même requête) :

- le montant à payer pour ce produit

1.A partir de la table Produit, afficher "Produit non disponible" lorsque
l'attribut Indisponible vaut 1, et "Produit disponible" sinon.

2.À partir de la table DetailsCommande, indiquer les infos suivantes en fonction de
la remise

si elle vaut 0 : "aucune remise"

si elle vaut entre 1 et 5% (inclus) : "petite remise"

si elle vaut entre 6 et 15% (inclus) : "remise modérée"

sinon :"remise importante"

3.Indiquer pour les commandes envoyées si elles ont été envoyées en retard (date d'envoi DateEnv supérieure (ou égale) à la date butoir ALivAvant) ou à temps

```python
#1.A partir de la table Produit, afficher "Produit non disponible" lorsque l'attribut Indisponible vaut 1, et "Produit disponible" sinon.
query = ''' SELECT nomProd, CASE WHEN (indisponible=1) THEN "Produit non disponible" 
ELSE "Produit disponible"
END AS informations
FROM produit ORDER BY indisponible'''

sql_to_df(query)
```

```python
#2.À partir de la table DetailsCommande, indiquer les infos suivantes en fonction de la remise
#si elle vaut 0 : "aucune remise"
#si elle vaut entre 1 et 5% (inclus) : "petite remise"
#si elle vaut entre 6 et 15% (inclus) : "remise modérée"
#sinon :"remise importante"

#ne pas utiliser BETWEEN 
query = ''' SELECT noCom, remise, CASE WHEN (remise=0) THEN "Aucune remise"
WHEN (remise <0.06) THEN "Petite remise"
WHEN (remise <0.16) THEN "Remise modérée"
ELSE "Remise importante"
END AS informations
FROM detailCommande '''

sql_to_df(query).head(20)
```

```python
#3.Indiquer pour les commandes envoyées si elles ont été envoyées en retard (date d'envoi DateEnv supérieure (ou égale) à la date butoir ALivAvant) ou à temps
#penser à toutes les éventualités (colis envoyé a temps en retard ou pas du tout)
query = ''' SELECT noCom, codeCli, CASE WHEN (dateEnv>=aLivAvant) THEN "En retard"
WHEN (dateEnv<aLivAvant) THEN "A temps"
ELSE "Non envoyé"
END AS informations
FROM commande '''

sql_to_df(query)
```

Dans une même requête, sur la table Client :
* Concaténer les champs Adresse, Ville, CodePostal et Pays dans un nouveau
champ nommé Adresse_complète, pour avoir : Adresse, CodePostal, Ville, Pays
* Extraire les deux derniers caractères des codes clients
* Mettre en minuscule le nom des sociétés
* Remplacer le terme "Owner" par "Freelance" dans Fonction
* Indiquer la présence du terme "Manager" dans Fonction

```python
query = ''' SELECT CONCAT(adresse,", ", codePostal,", ", ville,", ", pays) AS Adresse_complète, 
SUBSTR(codeCli,4,2) AS minCodecli, 
LOWER(societe) AS Societes,
REPLACE(fonction,"Owner","Freelance") AS Fonctions,
CASE WHEN (fonction LIKE "%manager%") THEN "OUI"
ELSE "NON"
END AS Manager
FROM clients '''

sql_to_df(query)
```
